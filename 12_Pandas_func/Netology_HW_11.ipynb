```python
"""Задание 1**
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class"""

import pandas as pd

movies_id = pd.read_csv('ratings.csv')                                   # читаем из файла ratings.csv
selected_movies_id = movies_id[['movieId', 'rating']]                    # выбираем 2 столбца для представления
movies_name = pd.read_csv('movies.csv')[['movieId', 'title']]            # читаем из файла movies.csv и выбираем нужные столбцы
joined = selected_movies_id.merge(movies_name, on='movieId', how='left') # соединяем данные таблиц по столбцу movieId

def level_ratings(param):                                                # пишем логику классификации фильмов 
    if param['rating'] <= 2:
        return 'Низкий рейтинг'
    elif 2 < param['rating'] <= 4:
        return 'Средний рейтинг'
    else: return 'Высокий рейтинг'

joined['class'] = joined.apply(level_ratings, axis=1)                   # создаем столбец class 
joined.head(10)                                                         # и применяем к нему написанную функцию
        
```
