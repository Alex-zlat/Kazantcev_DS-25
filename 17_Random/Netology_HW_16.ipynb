```python
"""
Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает
или теряет игрок за одну игру?

РЕШЕНИЕ:
Ответы: игра невыгодна (вкладывается в нее 100 рублей, доход - только 60,3675 рубля); за одну игру (партию) потеря 0,396325 руб. 
Расчетные величины
A - выпал выигрыш
Н1 - выпало 3 одинаковые цифры в диапазоне [111, 333, 555, 777, 999] (Р(Н1) = 0,001=(1/10)*(1/10)*(1/10))
Н2 - выпало 2 одинаковые цифры в диапазоне [00, 77] (Р(Н2) = 0,01=(1/10)*(1/10))
Н3 - выпала одна цифра в диапазоне [0, 7] (Р(Н3) = 0,1=(1/10))
P(A/H1)- вероятность, что выпала одна из выигрышных комбинаций, имеющая 3 одинаковые цифры заданного диапазона (0,5=5/10=1/2),
  т.е. из всех возможных XXX (000, 111, 222, 333, 444 и т.д.) ровно 50% заданных комбинаций 
P(A/H2)- вероятность, что выпала одна из выигрышных комбинаций, имеющая 2 одинаковые цифры заданного диапазона (0,2=2/10=1/5)
P(A/H3)- вероятность, что выпала одна из выигрышных комбинаций, имеющая 1заданную цифру установленного диапазона (0,2=2/10=1/5)
По формуле полной вероятности вероятность выигрыша: Р(А) = P(A/H1)*Р(Н1) + P(A/H2)*Р(Н2) + P(A/H3)*Р(Н3) =
                                                  =100*((0,5*0,001)+(0,2*0,01)+(0,2*0,1))=2,25%
В среднем игрок получает при выпадении Н1: (10+15+50+100+200)/5 = 75 руб. 
                                       Н2: (2+5)/2 = 3,5 руб.
                                       Н3: (1+3)/2 = 2 руб. 
Совокупно игрок выиграет в среднем: (75 +3,5 +2)/3 = 26,83 руб. 
Игрок заплатит 100 рублей за 100 попыток. Чтобы узнать сумму выигрыша, составим пропорцию:
1 процент - 26,83
2,25 процента - x. Отсюда со 100 рублей выиграть удастся (х) = 60,3675 руб. 
Потери со 100 партий: 100-60,3675 = 39,6325 руб. Потеря за одну игру (партию): 39,6325/100 = 0,396325 руб.
"""
import numpy as np

trials = 10000
successes=0
for i in range(trials):
    random_score = np.random.randint(0, 1000)
    if random_score ==111:
        successes +=10
    elif random_score ==333:
        successes +=15
    elif random_score ==555:
        successes +=50
    elif random_score ==777:
        successes +=200
    elif random_score ==999:
        successes +=100
    elif random_score ==333:
        successes +=10
    # если выпало *00:
    elif random_score  % 100 == 0:
        successes +=2
    # если выпало *77:
    elif (random_score - 77)  % 100 == 0:
        successes +=5
    # если выпало **7:
    elif (random_score - 7)  % 10 == 0:
        successes +=3
    # если выпало **0:
    elif random_score   % 10 == 0:
        successes +=1
p_monte_carlo = successes/trials
p = 1 - (39.6325/100)  # теоретическая вероятность
print('Накопленная сумма: ', successes,  '\nПотраченные деньги: ', trials, '\nСмоделированная вероятность: ', p_monte_carlo, '\nРасчетная вероятность : ', p)
```
