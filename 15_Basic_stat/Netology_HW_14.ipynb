```python
"""
Задание 1. Базовое изучение
Изучить представленный набор данных на основе описания его столбцов 
и выбрать 8 столбцов для дальнейшего изучения 
(среди них должны быть как числовые, так и категориальные). 
Провести расчет базовых метрик для них, кратко описать результаты"""

import numpy as np
import pandas as pd

horse_list = pd.read_csv('horse_data.csv', header = None, na_values='?')


horse_list.columns = (['surgery?/хирургия?',                    # 'Хирургия? (1 -да/2 -нет)',
                       'Age',                                   # 'Возраст (1-взрослая/2-молодая)', 
                        'Hospital Number',                      # 'Номер госпитализации (id)',
                       'rectal temperature/ректальная t',       # 'Ректальная температура (норм 37,8)',   
                        'pulse/пульс',                          # 'Пульс (норм 30-40)',
                       'respiratory rate/част.дыхания',         # 'Частота дыхания (норм 8-10)', 
                        'temperature of extremities',           # 'температура конечностей (1-норм, 2-горячо, 3-прохладно, 4-холодно)',
                       'peripheral pulse',                      # 'периферический пульс (1-норм/4-отсутствие)'
                        'mucous membranes',                     # 'слизистые оболочки (1-норм/6-темно-синий)',
                       'capillary refill time',                 # 'время пополнения капилляров (1 -до 3сек/2 -больше 3 сек)'
                        'pain/боль',                            # 'боль(1 -норм/5 -непрерывная сильная боль)',
                       'peristalsis',                           # 'перистальтика(1-гипермотиль, 2-норм, 3-гипо, 4-отсутствует)'
                        'abdominal distension',                 # 'вздутие живота(1-нет/4-сильно)',
                       'nasogastric tube',                      # 'назогастральный зонд(1 -нет/2 -существенно)'
                        'nasogastric reflux',                   # 'назогастральный рефлюкс(1-нет, 2-больше 1 л, 3-меньше 1 л)',
                       'nasogastric reflux PH',                 # 'назогастральный рефлюкс РН(3-4 - норм)'
                        'rectal examination',                   # 'ректальное исследование(1 -норм/4 -отсутствует)',
                       'abdomen',                               # 'брюшная полость(1-норм, 2-другое, 3--5-проблемы)'
                        'packed cell volume/объем упак.клеток', # 'объем упакованных клеток(30-50-норм)',
                       'total protein/общий протеин',           # 'общий белок(6-7,5-норм)'
                        'abdominocentesis appearance',          # 'внешний вид абдоминоцентеза(1-норм, 2 и 3-проблемы)',
                       'abdomcentesis total protein/абдоминоцентез общий белок', # 'абдоминоцентез общий белок(чем выше, тем хуже)'
                        'outcome',                              # 'гибель (1-жив, 2-мертв, 3-усыплен)',
                       'surgical lesion?',                      # 'хирургическое повреждение?(1 -да/2 -нет)'
                        'type of lesion',                       # 'тип повреждения(1-11-орган поражения/1-4-тип_смерти/1,2,0/0-10)', 
                       '25',                                    #'25',
                        '26',                                   # '26',
                       'cp_data'])                              #'cp_data (есть ли данные о патологии(да/нет))'
                                                  
                                                   
new_table = horse_list[['surgery?/хирургия?', 'rectal temperature/ректальная t', 'pulse/пульс', 'respiratory rate/част.дыхания', 'pain/боль', 'packed cell volume/объем упак.клеток', 'total protein/общий протеин', 'abdomcentesis total protein/абдоминоцентез общий белок']]
new_table.describe()
"""
Из общих параметров мы видим, что наименьшие пропуски данных у столбца проверки хирургического вмешательства. 
И больше всего пропусков в данных показателя "абдоминоцентез общий белок"(abdomcentesis total protein): 
фактически треть данных отсутствует.
Это значит, что результат при расчете стат.показателей может легко быть искажен, 
поэтому при замене данных необходимо внимательно подойти к методу замены. 
Категориальные показатели: "Хирургия?" (surgery?), "Боль"(pain), остальные числовые. При этом по категории боли 
среднее арифметическое (мат.ожидание) все же косвенно говорит о том, 
что обычно при обращении животные испытывают боль, когда их обследует ветеринар, 
нежели легкое недомогание. В этом смысле данные показатель условно можно
считать и числовым.
Отклонения индивидуальных значений показателей от стандартного отклонения есть мера рассеивания величины. 
По большинству показателей 
эта характеристика демонстрирует высокий уровень, что снижает достоверность данных, описываемых средним арифметическим значением.
Низкое рассеивание мы имеем лишь по показателям ректальной температуры (rectal temperature) и общего протеина (total protein). 
В остальных случаях стандартное отклонение 
говорит о том, что индивидуальные значения величины отклоняются от мат. ожидания примерно на 25-30 процентов. 
Высокий размах вариации демонстрируют характеристики пульса (pulse), частоты дыхания (respiratory rate), 
объема упакованных клеток (packed cell volume), общего протеина(total protein) , 
абдоминоцентеза общего белка (abdomcentesis total protein). Это может свидетельствовать о наличии выбросов."""
```

```python
"""Задание 2. Работа с выбросами
В выбранных числовых столбцах найти выбросы, 
выдвинуть гипотезы об их причинах и проинтерпретировать результаты. 
Принять и обосновать решение о дальнейшей работе с ними.
"""

# Исследуем выбросы по столбцам: pulse/пульс, respiratory rate/част.дыхания, packed cell volume/объем упак.клеток,
# total protein/общий протеин (см. далее)
# Кратко выводы: 
#   - pulse/пульс: пренебречь наличием выбросов (их присутствие не искажает результат)
#   - respiratory rate/част.дыхания: учитывать выбросы (их присутствие искажает результат)
#   - packed cell volume/объем упак.клеток: пренебречь наличием выбросов (их присутствие не искажает результат)
#   - abdomcentesis total protein/абдоминоцентез общий белок: учитывать выбросы (их присутствие искажает результат)

# ******  1  ******
# Вначале найдем и интерпретируем выбросы столбца pulse/пульс. Для этого на первом этапе рассчитываем значения, лежащие внутри 
# "нормального" диапазона:
q1_1 = new_table['pulse/пульс'].quantile(0.25)
q3_1 = new_table['pulse/пульс'].quantile(0.75)
iqr_1 = q3_1 - q1_1
lower_bound_1 = q1_1 - (1.5 * iqr_1) 
upper_bound_1 = q3_1 + (1.5 * iqr_1)
remove_outliers_1 = new_table[new_table['pulse/пульс'].between(lower_bound_1, upper_bound_1, inclusive=True)]

#  А теперь оставим только показатели выбросов по пульсу: 
out_1 = pd.concat([new_table, remove_outliers_1]).drop_duplicates(keep=False)

# рассчитаем среднее значение и медиану данных без выбросов: мы видим существенное различие в данных выбросов и мат.ожидания:
# "на глаз" это в 2 раза и выше. 
pulse_mean = remove_outliers_1['pulse/пульс'].mean()
pulse_median = remove_outliers_1['pulse/пульс'].median()

# Чтобы выяснить, насколько средний показатель столбца, включающего все данные вместе с выбросами,
# отличается от среднего без выброса, 
# рассчитаем между ними разницу (все данные минус данные без выбросов): 

diff_1 = new_table['pulse/пульс'].mean() - remove_outliers_1['pulse/пульс'].mean()
print(diff_1)

# Разница diff_1 нам говорит о том, что при значительных отклонениях значений выбросов от мат.ожидания, 
# они не очень существенно влияют на итог. По всей видимости, это связано с тем, что "девиантных" значений не столь много. 
# Поэтому в дальнейшем наличием выбросов можно пренебречь
```

```python
# ******  2  ******
# Теперь найдем и интерпретируем выбросы столбца частоты дыхания (respiratory rate). Для этого вначале рассчитываем значения, 
# лежащие внутри "нормального" диапазона:
q1_2 = new_table['respiratory rate/част.дыхания'].quantile(0.25)
q3_2 = new_table['respiratory rate/част.дыхания'].quantile(0.75)
iqr_2 = q3_2 - q1_2
lower_bound_2 = q1_2 - (1.5 * iqr_2) 
upper_bound_2 = q3_2 + (1.5 * iqr_2)
remove_outliers_2 = new_table[new_table['respiratory rate/част.дыхания'].between(lower_bound_2, upper_bound_2, inclusive=True)]

# Животные больные, и при их здоровом положении многие значения пришлось бы отсекать (сократить диапазон нормальности)
# (слишком много встречается значительных отклонений).
# Но полскольку речь идет о заболевших особях, то будем считать низкое дыхание допустимым значением, а данные корректными
# Далее оставим только показатели выбросов по частоте дыхания: 
out_2 = pd.concat([new_table, remove_outliers_2]).drop_duplicates(keep=False)


# Мы видим, что за пределами "нормального" диапазона достаточно много девиантных экстремумов частоты дыхания 
# (строк с пустыми значениями 64).

# Рассчитаем среднее значение и медиану данных без выбросов. 
pulse_mean = remove_outliers_2['respiratory rate/част.дыхания'].mean()
pulse_median = remove_outliers_2['respiratory rate/част.дыхания'].median()

# Мы видим, что выбросы существенно искажали картину. 
# В действительности частота дыхания больных животных далека от нормы (здоровый диапазон 30-40) 
# Проверим это утверждение математически. 
# Чтобы выяснить, насколько средний показатель столбца, включающего все данные вместе с выбросами,
# отличается от среднего без выброса, 
# рассчитаем между ними разницу (все данные минус данные без выбросов): 

diff_2 = new_table['respiratory rate/част.дыхания'].mean() - remove_outliers_2['respiratory rate/част.дыхания'].mean()
print(diff_2)

# Видим отклонение, и в дальнейшем имеет смысл учитывать при расчетах наличие девиантных выбросов (отсекать их)
```

```python
# Далее найдем и интерпретируем выбросы столбца объема упакованных клеток (packed cell volume). Для этого в первую очередь
# рассчитаем значения, лежащие внутри "нормального" диапазона:
q1_3 = new_table['packed cell volume/объем упак.клеток'].quantile(0.25)
q3_3 = new_table['packed cell volume/объем упак.клеток'].quantile(0.75)
iqr_3 = q3_3 - q1_3
lower_bound_3 = q1_3 - (1.5 * iqr_3) 
upper_bound_3 = q3_3 + (1.5 * iqr_3)
remove_outliers_3 = new_table[new_table['packed cell volume/объем упак.клеток'].between(lower_bound_3, upper_bound_3, inclusive=True)]

#  А теперь оставим только показатели выбросов по объему упакованных клеток: 
out_3 = pd.concat([new_table, remove_outliers_3]).drop_duplicates(keep=False)
out_3

# Мы видим, что за пределами "нормального" диапазона лежат существенные отклонения значений.

# Рассчитаем среднее значение и медиану данных без выбросов. 
pulse_mean = remove_outliers_3['packed cell volume/объем упак.клеток'].mean()
pulse_median = remove_outliers_3['packed cell volume/объем упак.клеток'].median()

# Наблюдаем несущественное отклонение. Получим точную цифру:

diff_3 = new_table['packed cell volume/объем упак.клеток'].mean() - remove_outliers_3['packed cell volume/объем упак.клеток'].mean()
print(diff_3)

# При нормальном диапазоне показателя 30-50 значение искажений от выбросов в 0.32 не столь велико. Им можно пренебречь.
```

```python
# Далее найдем и интерпретируем выбросы столбца абдоминоцентеза общего белка. Для этого в первую очередь
# рассчитаем значения, лежащие внутри "нормального" диапазона:
q1_4 = new_table['abdomcentesis total protein/абдоминоцентез общий белок'].quantile(0.25)
q3_4 = new_table['abdomcentesis total protein/абдоминоцентез общий белок'].quantile(0.75)
iqr_4 = q3_4 - q1_4
lower_bound_4 = q1_4 - (1.5 * iqr_4) 
upper_bound_4 = q3_4 + (1.5 * iqr_4)
remove_outliers_4 = new_table[new_table['abdomcentesis total protein/абдоминоцентез общий белок'].between(lower_bound_4, upper_bound_4, inclusive=True)]

#  А теперь оставим только показатели выбросов по абдоминоцентезу общего белка: 
out_4 = pd.concat([new_table, remove_outliers_4]).drop_duplicates(keep=False)
out_4

# Видим много пустых строк. Рассчитаем среднее значение и медиану данных без выбросов. 
pulse_mean = remove_outliers_4['abdomcentesis total protein/абдоминоцентез общий белок'].mean()
pulse_median = remove_outliers_4['abdomcentesis total protein/абдоминоцентез общий белок'].median()

# Отклонение имеется: аннулировав выбросы, мы на несколько сотых долей снижаем указанные показатели. 
# Уточним их математически:

diff_4 = new_table['abdomcentesis total protein/абдоминоцентез общий белок'].mean() - remove_outliers_4['abdomcentesis total protein/абдоминоцентез общий белок'].mean()
print(diff_4)

# учитывая, что чем выше рассматриваемый показатель, тем хуже для здоровья животного,
# то удаление выбросов дает более "здоровую" картину. Есть смысл откидывать их при расчете показателей в дальнейшем.
```
