```python
"""Уровень 0:
Посчитайте значение первой и второй производной функции:

f(x) = x**5 + 4sin(2x) + cos(3x+3)

В точке x = 1.
"""

from scipy.misc import derivative 
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
    return x**5 + 4*math.sin(2*x) + math.cos(3*x+3)

print(derivative(f, 1.0, dx=1e-6))
print(derivative(f, 1.0, dx=1e-6, n=2))
```

```python
"""Уровень 1:
Постройте график первой и второй производной функции:

f(x) = sin**5(2x+1)

На отрезке [-5; 5]
"""

def f1(x):
    return (np.sin**5)(2*x+1)

x=np.linspace(-10,10)
myfunction = np.vectorize(f)  # np.vectorize () работает как цикл for, 
                              #который позволяет переданной в нее функции принимать один элемент из массива
fx =myfunction(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f2x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative_1')
plt.plot(x, f2x, label='derivative_2')
plt.legend()
```
