```python
"""Задание.
В этом задании нужно убедиться, что ЦПТ действительно работает.

Создайте случайную величину из любого выбранного вами распределения, 
для разных значений n сгенерируйте 1000 выборок размера n 
и постройте гистрограммы средних этих выборок
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

rv = sts.t(5)
sample = rv.rvs(size=1000) # rv.rvs(size=1000) к распределению rv применим метод задания случайных чисел 
                           # .rvs(size=1000) в количестве 1000 экземпляров      
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)                                                   # .pdf(x) - метод плотности вероятности заданного распределения 
plt.hist(sample, density=True, bins=30, label='sample histogram') # density - плотность 
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, f(x)')                           # фракция образцов 
plt.xlabel('$x$')
plt.show()
```

```python
# Параметры запуска: размеры выборок и количество повторов генерации
n1 = 2
rv = sts.norm(loc=1, scale=3)

samples_count = 1000
# Массивы для сохранения средних по выборкам
a1 = np.array([])

for number in range(samples_count):
    # На каждом цикле генерируем выборки разных размеров
    sample1 = rv.rvs(size=n1)
    sample1.mean()
    # Считаем среднее по каждой выборке и добавляем в массив средних
    a1 = np.append(a1, sample1.mean())
```

```python
# Строим графики
# Подписи для графиков
label_1 = 'sample of ' + str(n1) + ' histogram'
plt.hist(a1, density=True, label=label_1)

plt.legend()
plt.ylabel('samples means')
plt.xlabel('$x$')
```
