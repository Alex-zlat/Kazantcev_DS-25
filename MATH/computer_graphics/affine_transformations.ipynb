```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
"""
Изобразим заданную фигуру на координатной сетке
"""
# Объявим матрицу, содержащую координаты фигуры 
A = [
[-2.5, -2.2, 1],  
[-1.8, 4, 1], 
[2, 1.5, 1],
[-1, -2.2, 1],
[-1.2, 0, 1],
[-2.5, -2.2, 1]
]
```

```python
# Теперь сформируем массив Numpy, которому передадим объявленную на предыдущем шаге матрицу А
A = np.array(A)
```

```python
# Соберем значения координат, по которым следом построим plot-визуализацию
x = A[:,0]   
y = A[:,1]
```

```python
# Оформляем визуализацию
fig, ax = plt.subplots()
# реализуем построение осей координат
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
# на плоскости изобразим график
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
plt.ylim([-5, 5])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(-1.3, 2.1, 'Исходная фигура',
        rotation = -24,
        fontsize = 12)
ax.set_title('Рис.1. Отправной график')
plt.show()
```

```python
"""
Сместим матрицу относительно ее центра в точке Q(x=-0.25, y=0.9) на данный вектор
"""

Q = np.array([
[1, 0, 0],
[0, 1, 0],
[0.25, -0.9, 1],
])
```

```python
vectorized_matrix = A@Q
x = vectorized_matrix[:,0]
y = vectorized_matrix[:,1]
```

```python
# Оформляем визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
plt.ylim([-5, 5])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(0.2, -3.5, 'Смещение\n центра',
        rotation = 35,
        fontsize = 12)
ax.set_title('Рис.2. График с фигурой,\n смещенной по своему центру')
plt.show()
vectorized_matrix
```

```python
"""
Осуществим масштабирование заданной фигуры
"""

# Определим масштабирующую матрицу 
B = np.array([
[2, 0, 0],
[0, 2, 0],
[0, 0, 2],
])
```

```python
# Масштабирование осуществим через перемножение матриц 
reduce_2_times = A@B       # reduce_2_times = np.dot(A,B)
x = reduce_2_times[:,0]
y = reduce_2_times[:,1]
```

```python
# Оформляем визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()

plt.ylim([-9, 9])
plt.xlim([-9, 9])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(-1, -6, 'Масштабированная\n фигура',
        rotation = 37,
        fontsize = 12)
ax.set_title('Рис.3. График с фигурой,\n в 2 раза увеличенной')
plt.show()
reduce_2_times
```

```python
"""
Осуществим поворот заданной фигуры на 130 градусов
"""
# объявим матрицу поворота
C = np.array([
[math.cos(math.radians(100)), math.sin(math.radians(100)), 0],
[math.sin(math.radians(100)), math.cos(math.radians(100)), 0],
[0,0,1]
])
```

```python
rotation_matrix = A@C  
```

```python
# Соберем значения координат, по которым следом построим plot-визуализацию
x = rotation_matrix[:,0]
y = rotation_matrix[:,1]
```

```python
# Оформляем визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
# Определим диапазон сетки графика
plt.ylim([-5, 5])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(1.3, -1.3, 'Развернутая фигура',
        rotation = -42,
        fontsize = 12)
# Добавляем заголовок
ax.set_title('Рис.4. График с исходной фигурой,\n повернутой на 100 градусов по обеим осям')
plt.show()
rotation_matrix
```

```python
"""
Осуществим перенос заданной фигуры 
"""

# Определим матрицу переноса 
D = np.array([
[1, 0, 0],
[0, 1, 0],
[2, 2, 1],
])
```

```python
transfer_matrix = A@D  
```

```python
# Соберем значения координат, по которым следом построим plot-визуализацию
x = transfer_matrix[:,0]
y = transfer_matrix[:,1]
```

```python
# Оформляем визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
# Определим диапазон сетки графика
plt.ylim([-5, 9])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(0.5, 4.2, 'Перенесенная фигура',
        rotation = -19,
        fontsize = 12)
# Добавляем заголовок
ax.set_title('Рис.5. График с перенесенной исходной фигурой')
plt.show()
transfer_matrix
```

```python
"""
Осуществим отражение заданной фигуры 
"""

# Определим отражающую матрицу 
E = np.array([
[1, 0, 0],
[0, -1, 0],
[0, 0, 1],
])
```

```python
# Сформируем расчет для отраженной матрицы 
reflection_matrix = A@E  
x = transfer_matrix[:,0]
y = transfer_matrix[:,1]
```

```python
# Оформим визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
# Определим диапазон сетки графика
plt.ylim([-5, 9])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(0.5, 4.2, 'Отраженная фигура',
        rotation = -20,
        fontsize = 12)
# Добавляем заголовок
ax.set_title('Рис.6. График с отраженной исходной фигурой')
plt.show()
reflection_matrix
```

```python
"""
Выполним одномоментно все преобразования матрицы
"""
# Объявим матрицу, содержащую координаты фигуры 
A1 = [
[-2.5, -2.2, 1],  
[-1.8, 4, 1], 
[2, 1.5, 1],
[-1, -2.2, 1],
[-1.2, 0, 1],
[-2.5, -2.2, 1]
]

transformation_matrix = A@Q@B@C@D@E
# transformation_matrix = A1@B
x = transfer_matrix[:,0]
y = transfer_matrix[:,1]
```

```python
# Оформим визуализацию
fig, ax = plt.subplots()
ax.axhline(y=0, color='k', linewidth = 5)
ax.axvline(x=0, color='k', linewidth = 5)
ax.plot(x, y, color='green', linewidth = 4)
#  Включим второстепенные деления осей
ax.minorticks_on()
# Определим диапазон сетки графика
plt.ylim([-5, 9])
plt.xlim([-5, 5])
# определим внешний вид основной сетки
ax.grid(which='major',color = 'k', linewidth = 1)
#  Определяем внешний вид линий вспомогательной сетки
ax.grid(which='minor', color = 'k', linestyle = ':')
# добавим подпись к графику
#  Добавляем текст:
ax.text(0.5, 4.2, 'Измененная фигура',
        rotation = -20,
        fontsize = 12)
# Добавляем заголовок
ax.set_title('Рис.6. График всех преобразований над фигурой')
plt.show()
transformation_matrix
```
