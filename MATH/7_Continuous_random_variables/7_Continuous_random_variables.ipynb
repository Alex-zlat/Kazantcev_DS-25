```python
"""Реализация задачи через циклы
"""

for vals in [(2.0, 0.5), (1.5, 0.5), (2.5, 0.5), (2.0, 0.45), (2.0, 0.55)]:
    norm_rv = sts.norm(loc=vals[0], scale=vals[1])
    x = np.linspace(0, 4, 100)
    y = norm_rv.pdf(x)
    plt.plot(x, y, label=f"µ = {vals[0]}, σ = {vals[1]}")
    
plt.title("Нормальное рапределение")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```

```python
for lamda in [0.1, 0.5, 0.9, 1.1, 1.5]:
    expon_rv = sts.expon(lamda)
    x = np.linspace(0, 5, 1000)
    y = expon_rv.pdf(x)
    plt.plot(x, y, label=f"λ = {lamda}")
plt.title("Экспоненциальное рапределение")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```

```python
for n in [1, 2, 3, 5, 30]:
    t_rv = sts.t(n)
    x = np.linspace(-5, 5, 1000)
    y = t_rv.pdf(x)
    plt.plot(x, y, label=f"n = {n}")
    
plt.title("Распределение Стьюдента")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```

```python
"""Задание 2
Непрерывная случайная величина задана плотностью распределения: f(ξ) = C, ξ ∈ [0;5]; f(ξ) = 0, ξ notin [0;5]
Найдите C, математическое ожидание ξ а также вероятность попадания в отрезок [3.5;7], т.е. P(3.5 < ξ < 7).
"""

a=0
b=5
uniform = sts.uniform(a, b-a)
C = uniform.pdf(3)
print('C: ', C)
print('Мат.ожидание: ', uniform.mean())
print('Вероятность попадания в отрезок [3,5; 7]: ', uniform.cdf(7)-uniform.cdf(5))
```
