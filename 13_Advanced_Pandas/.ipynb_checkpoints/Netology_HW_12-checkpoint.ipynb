```python
"""Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic
для источников paid и email из России - ставим ad
для источников paid и email не из России - ставим other
все остальные варианты берем из traffic_source без изменений"""

import pandas as pd

log = pd.read_csv('visit_log.csv', sep=';')
def modified_table(row):
    """
    Функция определяет значения столбца source_type в зависимости от значений столбцов traffic_source и region 
    """
    if 'yandex' in row['traffic_source']:
        return row['traffic_source'].replace('yandex', 'organic')  
    elif 'google' in row['traffic_source']:
        return row['traffic_source'].replace('google', 'organic')
    elif 'paid' in row['traffic_source'] and 'Russia' in row['region']:
        return row['traffic_source'].replace('paid', 'ad')
    elif 'paid' in row['traffic_source'] and 'Russia' not in row['region']:
        return row['traffic_source'].replace('paid', 'other')
    else:
        return row['traffic_source']

log['source_type'] = log.apply(modified_table, axis = 1)
log.head()
```

```python
"""Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
Под временем жизни понимается разница 
между максимальным и минимальным значением 
столбца timestamp для данного значения userId"""

import pandas as pd

ratings = pd.read_csv('ratings.csv')

df_ratings_agg = (ratings[['userId', 'rating', 'timestamp']]                        # выбираем для вывода столбцы
                  .groupby(by='userId')                                             # группируем строки по столбцу пользователя
                 .agg({'rating': 'count', 'timestamp': ['min', 'max']}))            # задаем расчет 3-х функций для столбцов

df_ratings_agg.loc[df_ratings_agg['rating']['count'] > 100]                         # выбираем строки с оценками больше 100

df_ratings_agg['average_user_life_cicle'] = (df_ratings_agg['timestamp']['max']     # в новый столбец вычисляем срок жизненного   
                                             -df_ratings_agg['timestamp']['min'])   # цикла каждого пользователя
average_time = (df_ratings_agg
                .average_user_life_cicle                                            # к новому столбцу применяем расчет
                .mean())                                                            # среднего значения цикла 
print(round(average_time))                                                          # округляем и выводим на экран  

```
