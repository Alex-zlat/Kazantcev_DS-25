```python
"""Обязательная часть
Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. 
Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>"""

import requests
from bs4 import BeautifulSoup 
import time
import pandas as pd
import numpy as np

URL = 'https://habr.com/ru/search/?q=python#h'
KEYWORDS = ['python', 'парсинг']


habr_posts = pd.DataFrame()
req = requests.get(URL)
time.sleep(0.3)
soup = BeautifulSoup(req.text, 'html.parser')
posts = soup.find_all('article', class_='post post_preview')
    
for post in posts:
    for word in  KEYWORDS:
        if word in post.text.lower():
            article_date = post.find('span', class_='post__time').text
            article_title = post.find('a', class_='post__title_link').text
            article_text = post.find('div', class_='post__text').text
#             href_element = article_title.get('href')
            href_element = post.find('a', class_='post__title_link').get('href')
    
            row = {'Дата': article_date, 'Заголовок': article_title, 'Статья': article_text, 'Ссылка': href_element}
            habr_posts = pd.concat([habr_posts, pd.DataFrame([row])]) 


habr_posts.index = np.arange(len(habr_posts))
habr_posts
```

```python
"""Задание 2.
Обязательная часть
Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. 
Список email-ов задаем переменной в начале кода:
EMAIL = [zlat.tv@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>

Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы"""


import requests
import time
import pandas as pd
import json
import re

time_to_sleep_when_captcha = 5
params = {'emailAddresses': ['zlat.tv@mail.ru', 'zr-zlat@yandex.ru']}
url = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'
headers_ = {'Vaar-Version': '0', 'Vaar-Header-App-Product': 'hackcheck-web-awast'}

res = requests.post(url, data=json.dumps(params), headers=headers_)
# time.sleep(1.0)
res
```

```python
result = res.json()
my_list = []
pars_data = []
for email in result['summary']:
    try:
        for col_id in result['summary'][email]['breaches']:
            describe = result['breaches'][str(col_id)]['description']
            source = result['breaches'][str(col_id)]['site']
            date = re.findall(r'\d{3}', describe)[0]
            my_list.append(email)
            my_list.append(date)  
            my_list.append(source)
            my_list.append(describe)
            pars_data.append(my_list)
    except:
        sleep(time_to_sleep_when_captcha)
        time_to_sleep_when_captcha += 1
pars_data
                          
leaked_addresses = pd.DataFrame(pars_data, columns= ['почта', 'дата утечки', 'источник утечки', 'описание утечки'])
leaked_addresses

```
